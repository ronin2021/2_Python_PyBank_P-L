```python
"""
Python - Homework: PyBank

This script will use the Pathlib library to set the file path,
use the csv library to read in the file, iterate over each
row of the file to find/calculate the total number of months, net profit/loss, 
average change of changes in profit/loss over the entire data period, the
greatest increase in profits (month-year and amount) over the data period, and the
greatest increase in losses (month-year and amount) over the data period.

"""
# import the pathlib and csv library
from pathlib import Path
import csv

# set the file path
csvpath = Path('budget_data.csv')

# initialize lists to hold budget and month_years for futher use
budgets = []
month_years = []

# initialize line_number variable which would be used to move from line to line (row to row) while reading the csv file
line_number = 0

# open the input path as a file object with read
with open(csvpath, 'r') as csvfile:

    # pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter = ',')
   
    # go to the next row from the start of the file
    header = next(csvreader)

    # (which is often the first row/header) and iterate line_number by 1
    line_number += 1   
    
    # read each row of data after the header
    for row in csvreader:
        # set date variable equal to the value in the 0th column of each row
        month_year = str(row[0])
        # set budget variable equal to the value in the 1st column of each row
        budget = int(row[1])
        # Append the row date value to the list of month_years
        month_years.append(month_year)
        # Append the row budget value to the list of budgets
        budgets.append(budget)

# initialize variables to be used
number_of_months = 0
net_profit_loss = 0
change_in_profits = 0
total_change_in_profits_for_the_data_period = 0
average_change_in_profits_for_the_data_period = 0
greatest_increase_in_profits = 0
greatest_increase_in_profits_month_year = " "
greatest_increase_in_losses = 0
greatest_increase_in_losses_month_year = " "

# calculate the number of months, net profit/loss, average change in profit/loss, 
# greatest increase in profits, and greatest decrease in losses

# initializing the counter for the while loop
counter = 0

while counter < len(budgets):
    
    # increase the number of months by one per interation
    number_of_months += 1
    # summing up the net profit/loss by iteratively adding profit/loss each month 
    net_profit_loss += budgets[counter]
    
    # logic to determine average change in profit/loss
    
    # if the counter == 0 just increase the counter as there is NO prior data (no counter -1) in the budget[] list to calculate change in profits
    if counter < 1:
        counter += 1
        
    # if the counter > 0 uses the prior data (counter -1) in the budget[] list to calculate change in profits
    else:
        change_in_profits = budgets[counter] - budgets[counter - 1]
        #summing up the total change in profits for the data period
        total_change_in_profits_for_the_data_period += change_in_profits 
        
        # if change in profits >=0 and > greatest increase in profits, greatest increase in profits = change in profits 
        # i.e. greatest increase in profits gets updated and the greatest increase in profits month-year gets updated to the 
        # corresponding new month-year
        if change_in_profits >= 0:
            if change_in_profits > greatest_increase_in_profits:
                greatest_increase_in_profits = change_in_profits
                greatest_increase_in_profits_month_year = month_years[counter]
        
        # if change in profits <0 and < greatest increase in losses, greatest increase in losses = change in profits 
        # i.e. greatest increase in losses gets updated and the greatest increase in losses month-year gets updated to the 
        # corresponding new month-year
        else: 
            if change_in_profits < greatest_increase_in_losses:
                greatest_increase_in_losses = change_in_profits
                greatest_increase_in_losses_month_year = month_years[counter]
        counter +=1
        
# calculate the average change in profits for the data period, round to the nearest 2 decimal places
average_change_in_profits_for_the_data_period = round(total_change_in_profits_for_the_data_period / (number_of_months - 1), 2)

# print the metrics with a comma seperator for thousands
print("Financial Analysis")
print("---------------------------------------------------------")
print(f"Total Months: {number_of_months}")
print(f"Total: ${net_profit_loss: ,}")
print(f"Average Change: $ {average_change_in_profits_for_the_data_period: ,}")
print(f"Greatest Increase in Profits: {greatest_increase_in_profits_month_year} (${greatest_increase_in_profits: ,})")
print(f"Greatest Increase in Losses: {greatest_increase_in_losses_month_year} (${greatest_increase_in_losses: ,})")

# set the output file path
output_path = Path('output.txt')

# open the output path as a file object with write
with open(output_path, 'w') as textfile:
    # Write a line to the output file
    textfile.write("Financial Analysis\n")
    # Write a line to the output file
    textfile.write("--------------------------------------------------------------------------\n")
    # Write a line to the output file
    textfile.write(f"Total Months: {number_of_months}\n")    
    # Write a line to the output file
    textfile.write(f"Total: ${net_profit_loss: ,}\n")
    # Write a line to the output file
    textfile.write(f"Average Change: $ {average_change_in_profits_for_the_data_period:,}\n")
    # Write a line to the output file
    textfile.write(f"Greatest Increase in Profits: {greatest_increase_in_profits_month_year} (${greatest_increase_in_profits: ,})\n")
    # Write a line to the output file
    textfile.write(f"Greatest Decrease in Losses: {greatest_increase_in_losses_month_year} (${greatest_increase_in_losses: ,})")
```
