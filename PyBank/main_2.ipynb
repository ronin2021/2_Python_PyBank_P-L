```python
"""Homework 1: CSV Reader.

This script will use the Pathlib library to set the file path,
use the csv library to read in the file, iterate over each
row of the file to find/calculate the total number of months, net profit/loss, 
average change of changes in profit/loss over the entire period, the
greatest increase in profits (date and amout) over the period, and the
greatest decrease in profits (date and amout) over the period.
"""
```

```python
# import the pathlib and csv library
from pathlib import Path
import csv
```

```python
# set the file path
csvpath = Path('budget_data.csv')
```

```python
# initialize variables to hold month_year and budget
budgets = []
month_years = []
```

```python
# initialize line_number variable which would be used to move from line to line
line_number = 0
```

```python
# open the input path as a file object
with open(csvpath, 'r') as csvfile:
    # print the datatype of the file object
    print(type(csvfile))

    # pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter = ',')
    # print the datatype of the csvreader
    print(type(csvreader))

    # go to the next row from the start of the file
    header = next(csvreader)
    
    # print the header
    print(f"{header} <---- HEADER")

    # (which is often the first row/header) and iterate line_number by 1
    line_number += 1   
    
    # read each row of data after the header
    for row in csvreader:
        # print the row
        print(row)
        # set date variable equal to the value in the 0th column of each row
        month_year = str(row[0])
        # set budget variable equal to the value in the 1st column of each row
        budget = int(row[1])
        # Append the row date value to the list of dates
        month_years.append(month_year)
        # Append the row budget value to the list of budgets
        budgets.append(budget)
```

```python
# initialize variables
number_of_months = 0
net_profit_loss = 0
change_in_profits = 0
total_change_in_profits_for_the_period = 0
average_change_in_profits_for_the_period = 0
greatest_increase_in_profits = 0
greatest_increase_in_profits_date = " "
greatest_decrease_in_losses = 0
greatest_decrease_in_losses_date = " "

# calculate the number of months, net profit/loss, average change in profit/loss, 
# greatest increase in profits, and greatest decrease in losses
counter = 0

while counter < len(budgets):
    
    # sum the number of months
    number_of_months += 1
    # calculate the net profit/loss
    net_profit_loss += budgets[counter]
    
    # logic to determine average change in profit/loss
    if counter < 1:
        counter += 1
    else:
        change_in_profits = budgets[counter] - budgets[counter - 1]
        total_change_in_profits_for_the_period += change_in_profits 
        if change_in_profits >= 0:
            if change_in_profits > greatest_increase_in_profits:
                greatest_increase_in_profits = change_in_profits
                greatest_increase_in_profits_date = month_years[counter]
        else: 
            if change_in_profits < greatest_decrease_in_losses:
                greatest_decrease_in_losses = change_in_profits
                greatest_decrease_in_losses_date = month_years[counter]
        counter +=1
        
# Calculate the average change in profits for the period, round to the nearest 2 decimal places
average_change_in_profits_for_the_period = round(total_change_in_profits_for_the_period / (number_of_months - 1), 2)

# Print the metrics
print("Financial Analysis")
print("---------------------------------------------------------")
print(f"Total Months: {number_of_months}")
print(f"Total: ${net_profit_loss: ,}")
print(f"Average Change: $ {average_change_in_profits_for_the_period: ,}")
print(f"Greatest Increase in Profits: {greatest_increase_in_profits_date} (${greatest_increase_in_profits: ,})")
print(f"Greatest Decrease in Losses: {greatest_decrease_in_losses_date} (${greatest_decrease_in_losses: ,})")
```

```python
# Set the output file path
output_path = Path('output.txt')

# Open the output path as a file object
with open(output_path, 'w') as textfile:
    # Write a line to the output file
    textfile.write("Financial Analysis\n")
    # Write a line to the output file
    textfile.write("--------------------------------------------------------------------------\n")
    # Write a line to the output file
    textfile.write(f"Total Months: {number_of_months}\n")    
    # Write a line to the output file
    textfile.write(f"Total: ${net_profit_loss: ,}\n")
    # Write a line to the output file
    textfile.write(f"Average Change: $ {average_change_in_profits_for_the_period:,}\n")
    # Write a line to the output file
    textfile.write(f"Greatest Increase in Profits: {greatest_increase_in_profits_date} (${greatest_increase_in_profits: ,})\n")
    # Write a line to the output file
    textfile.write(f"Greatest Decrease in Losses: {greatest_decrease_in_losses_date} (${greatest_decrease_in_losses: ,})")
    
```
